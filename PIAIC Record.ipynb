```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('./Dataset/AIC.csv')
data2=pd.read_csv('./Dataset/CNC.csv')
data3=pd.read_csv('./Dataset/BCC.csv')
data=pd.concat([data1,data2,data3])
data.head()
```

# Calculating Total Enrollment

```python
enroll = [data1['Roll Number'].count(),data2['Roll Number'].count(),data3['Roll Number'].count()]
course_name = ['AI','CNC','BC']
plt.bar(course_name,enroll,width=0.3,color=['red','yellow','green'],alpha=0.8,ec='black')
plt.title('Number of Student Enrolled in the course',fontsize=15)
```

# Percentage Enrollment

```python
total = str(data1['Roll Number'].count()+data2['Roll Number'].count()+data3['Roll Number'].count())
colors = ['#FF3333', '#E4F031', '#71F660']
explode = (0.6, 0, 0, 0,0,0)  # explode 1st slice
plt.pie(enroll,colors=colors,autopct='%1.0f%%', shadow=False, startangle=90)
plt.legend(course_name,title="Total "+total)
plt.axis('equal')
plt.tight_layout()
plt.title('Percentage Enrollment in Course',fontsize=15)
plt.show()
```

# Calculating Percentile of Each Course

```python
f,a = plt.subplots(1,3)
a[0].hist(data1['Overall Percentile'],color='red',rwidth=2,ec='black',alpha=0.8)
a[1].hist(data2['Overall Percentile'],color='yellow',rwidth=2,ec='black',alpha=0.8)
a[2].hist(data3['Overall Percentile'],color='green',rwidth=2,ec='black',alpha=0.8)
a[0].set_title('AI PERCENTILE GRAPH',fontsize=15)
a[1].set_title('CNC PERCENTILE GRAPH',fontsize=15)
a[2].set_title('BCC PERCENTILE GRAPH',fontsize=15)
f.subplots_adjust(left=0, bottom=0, right=2, top=1, wspace=0.5, hspace=0.5)
```

# Working on Grades

```python
AI_grade = data1.groupby('Grade')
CNC_grade = data2.groupby('Grade')
BCC_grade = data3.groupby('Grade')
```

# Calculating AI Grades

```python
AI_grade_A = AI_grade.get_group('A')['Grade'].count()/enroll[0]*100
AI_grade_B = AI_grade.get_group('B')['Grade'].count()/enroll[0]*100
AI_grade_C = AI_grade.get_group('C')['Grade'].count()/enroll[0]*100
AI_grade_F = AI_grade.get_group('F')['Grade'].count()/enroll[0]*100
```

# Calculating CNC Grades

```python
CNC_grade_A = CNC_grade.get_group('A')['Grade'].count()/enroll[1]*100
CNC_grade_B = CNC_grade.get_group('B')['Grade'].count()/enroll[1]*100
# CNC_grade_C = CNC_grade.get_group('C')['Grade'].count()/enroll[1]*100
CNC_grade_F = CNC_grade.get_group('F')['Grade'].count()/enroll[1]*100
```

# Calculating BCC Grades

```python
BCC_grade_A = BCC_grade.get_group('A')['Grade'].count()/enroll[2]*100
BCC_grade_B = BCC_grade.get_group('B')['Grade'].count()/enroll[2]*100
BCC_grade_C = BCC_grade.get_group('C')['Grade'].count()/enroll[2]*100
BCC_grade_F = BCC_grade.get_group('F')['Grade'].count()/enroll[2]*100
```

```python
enroll = [AI_grade_A,CNC_grade_A,BCC_grade_A]
course_name = ['AI','CNC','BC']
plt.bar(course_name,enroll,width=0.3,color=['red','yellow','green'],alpha=0.8,ec='black')
plt.xlabel('Course Name',fontsize=15)
plt.ylabel('Percentage',fontsize=15)
plt.title('Comparison of A grade',fontsize=15)
```
